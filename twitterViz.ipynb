```python
'''
This cell loads the data from the .json file and stores it into:
 -- a list of dictionaries called tweetsList
 -- a pandas DataFrame called tweetsDF

You can assume that this cell will be executed before any of the visualization cells.

The cell includes the necessary imports for working with MatplotLib, and Seaborn
(as well as pandas and JSON files), but you can add others here as needed.

This cell will also display for use as a reference:
 -- the first tweet from tweetsList
 -- the first five tweets from tweetsDF
 
(This cell will also take a little while to run...)
'''

import json
import matplotlib.pyplot as plt
import seaborn as sns
import pprint
import pandas as pd

sns.set()

# Load dataset from the .json tweets file
with open('covidTweets.json', 'r', encoding="utf-8") as infile:
    tweetsList = json.load(infile)

# Convert the dataset into a partially-flattened pandas DataFrame   
tweetsDF = pd.json_normalize(tweetsList, sep='_')

# Sample output
print("The first tweet from tweetsList is:\n")
pprint.pprint(tweetsList[0])    

print("\n\nThe first five tweets from tweetsDF are:")
tweetsDF.head()
```

### Visualization 1

- **Question:** What are the most common key words used across all of the COVID-19 pandemic Tweets in recorded in the dataset (not including the keyword 'covid')?


- **Chart Type:**
   I constructed a **bar plot**, using matplotlib, for the purpose of showcasing the top 20 most frequent terms present across all of the tweets in the provided
              COVID-19 twitter dataset. The plot includes the most frequent terms not including the term 'covid'. A barplot seemed appropriate for this
   situation, as although we are examining frequencies, a measure commonly illustrated by histograms, barplots can provide a more descriptive
  comparison between the several different terms and their respective occurances

- **User Takeway(s):**
       Accumulating data regarding the most common words used across a rather nation-wide distressful time period, such as COVID-19, on a prominent social media platform such as Twitter, can
       convey the overall tone and thoughts of the people across the globe.
      Based on the barplot, the term 'people' seemed to have been the most frequent term across all of the recorded tweets, indicating that individuals largely emphasized the overall struggles of the human race during this period of time rather than choosing to convey their individual struggles during the pandemic. On a rather surprising note, it seems that
      some of the most frequent terms, despite attempting to cleanse the tweets of non-english words, were spanish terms such as 'contra'
      and 'para', indicating that spanish users may have been just as vocal as american users in social media outlets during COVID in 2022.

```python
# 1. Extract tweets from text column of dataframe

tweets = tweetsDF.text # tweets are stored in the text column of the dataframe

# 2. Data Cleaning

import nltk # ntlk library stands for Natural Language Toolkit. It provides numerous modules, functions and methods to process and parse our
            # strings for appropriate comparisons in computing.

tokenizer = nltk.RegexpTokenizer(r"\w+") # punctuation tokenizer

from nltk.corpus import stopwords

stopwords = stopwords.words('english') # stopwords are words that do not necessarily add subtance to a tweet, hence, it's better off to remove them
                                       # from the tweets when measuring term frequencies.

nltk.download('words')
english_words = set(nltk.corpus.words.words()) # the ntlk library contains a list of english words we can load for comparisons.

tweets = tweets.str.lower() # series.str.lower() transforms all the contents of the series to lowercase

def cleanse(tweet):
    tweet = tokenizer.tokenize(tweet) # removing punctuation from the tweet
    tweet = [term for term in tweet if term.isalpha()] # removing any possible numeric entries or symbols in the tweet
    tweet = [term for term in tweet if term not in stopwords] # removing stop words (as stop words can distort the term frequencies)
    tweet = [term for term in tweet if term in english_words and len(term) > 2] # removing non-english words
    return tweet
    
tweets = tweets.map(lambda tweet: cleanse(tweet)) # applying the tweet cleansing function to each tweet in the series through the use of map and lambda

from collections import Counter

T20 = Counter(tweets.explode()).most_common(21) # collections.Counter creates a dictionary of key-value pairs, where the keys are the unique elements of a
                                                # a collection, and the values are the corresponding frequencies of each term.

# tweets.explode() allows us to create a row for each term in each tweet of the series.

# it is fairly obvious, upon viewing the term frequencies, that 'covid' is the most frequent term, so we
# collect the top 21 most frequent terms in order to get the top 20 most frequent terms that do NOT
# include 'covid'

# 3. Plotting

T20 = pd.DataFrame(T20)

T20.columns = ['Terms', 'Frequency'] # titling the columns

T20 = T20[T20.Terms != 'covid'] # We are aiming to find the top 20 most frequent terms aside from 'covid', so it makes sense the filter the dataframe
                                # to not include the term 'covid'

T20.plot.bar(x='Terms', y='Frequency') # initializing bar chart where x-values are the terms and y-values are the respective frequencies

plt.ylabel("Frequency")

plt.title('Top 20 Most Frequent Terms Across COVID Tweets (2022)')
```

### Visualization 2

- **Question:** What is the distribution of the languages a tweet was tweeted in, amongst the 1000 longest **non-english** tweets?

- **Chart Type:** A **histogram** is ideal for showcasing distributions and frequencies of a given variable from a dataset. Here, we can use a histogram from seaborn to
                  model the frequencies and overall distribution of the recorded language of tweets amongst the 1000 longest **non-english** tweets in the
                  dataset. 


- **User Takeway(s):** After filtering the 1000 longest non-english tweets of the dataset, it seems that french (abbreviated as fr in the dataset) speaking tweeters were surprisingly prevelant during the later portion of the COVID-19 pandemic (2022), with portugese (pr) and spanish (es) speaking tweeters trailing closely. This shows that French tweeters were quite vocal across the pandemic when it came to social media outlets.

```python
tweetsDF['length'] = tweetsDF.text.map(lambda tweet: len(tweet.split())) # Here, we create a new column named length, for the purpose of recording the
                                                                         # length of each tweet based on the wordcount

NonEnglish = tweetsDF[tweetsDF['lang'] != 'en'] # extracting all Non-English tweets
Top1000 = NonEnglish[['length', 'lang']].sort_values(by='length', ascending=False).iloc[0:1000, 1:] # obtaining top 1000 longest tweets

sns.histplot(x='lang',data=Top1000) # plotting histogram of langs using seaborn's histogram plot function

# Appropriately enforcing labels and titles
plt.xlabel('Language (abbreviations)')
plt.ylabel('Frequency Across Tweets')
plt.title('Tweet Language Distribution Across 1000 Longest Non-English Tweets')
```

### Visualization 3

- **Question:** At what times throughout the days of COVID, recorded from the dataset, was tweeting activity the highest?

- **Chart Type:** We can use a swarmplot from seaborn to model the distribution of the top 20 most frequent times to tweet. This way, we can get an
              indicator as to when throughout a given day, tweeting activity was highest

- **User Takeway(s):** Based on the swarmplot, it seems that COVID tweeting activity peaks around hour 1 (datetime format), while being roughly evenly distributed throughout the rest of the day. This brings a bit of surprise, as it would be assumed that tweeting activity doesn't just follow a roughly equal distribution throughout a given day, but rather, varies randomly, especially considering the number of recorded days in the dataset. Aside fom a singular peak, tweeting activity remains relatively constant throughout the course of a given day.

```python
tweetsDF['Time'] = tweetsDF.timestamp.map(lambda timestamp: timestamp[timestamp.index(':')-2 : timestamp.index(':')+7].strip())

counts = Counter(tweetsDF.Time).most_common(20)

counts = [val[0] for val in counts]

Top20Times = tweetsDF[tweetsDF['Time'].isin(counts)]

sns.catplot(kind='swarm', x='Time',data=Top20Times), plt.xticks(rotation=90)

ax = plt.gca()

ax.set_title('Daily COVID Tweeting Activity (Top 20 Most Frequent Times)')
ax.set_xlabel('Recorded Times (Military Time)', labelpad=20)
```

### Visualization 4

- **Question:** Is there an association between the number of followers a twitter user has and the number of tweets (including retweets) they have under their account?
  
- **Chart Type:** A scatterplot from seaborn can effectively model the relationship between the number of tweets present under a given user's account (appears as user_statuses_count in the twitter API dataset) and the recorded number of followers they have listed in the dataset.

- **User Takeway(s):** Interestingly, there does not appear to be a clear or strong relationship between the number of followers a given twitter user has, and the number of tweets they have posted under their account. Furthermore, upon inspection of whether the accounts were verified or not, it seems that while verified twitter users may possess more twitter followers, on average, compared to unverified users, they do not have substantially more tweets under their accounts compared to unverified twitter users.

```python
tweetsDF.rename(columns = {'user_verified':'Verified'}, inplace = True) 

sns.relplot(x='user_statuses_count', y='user_followers_count', hue='Verified', data=tweetsDF)
# Here, we also color code the twitter user datapoints based on whether the user is verified on twitter or not.

ax = plt.gca()

ax.set_title('Number of Tweets Under Account VS Number of Followers of Account')
ax.set_xlabel('Number of Tweets Under Account (Millions)', labelpad=20)
ax.set_ylabel('Number of Followers (Ten Millions)', labelpad=20)
```
